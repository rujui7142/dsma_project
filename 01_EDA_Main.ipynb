```python
df_clean = df_sample.copy()

df_clean["tpep_pickup_datetime"] = pd.to_datetime(df_clean["tpep_pickup_datetime"])
df_clean["tpep_dropoff_datetime"] = pd.to_datetime(df_clean["tpep_dropoff_datetime"])

df_clean["trip_duration_min"] = (
    df_clean["tpep_dropoff_datetime"] -
    df_clean["tpep_pickup_datetime"]
).dt.total_seconds() / 60

df_clean["pickup_hour"] = df_clean["tpep_pickup_datetime"].dt.hour
df_clean["pickup_dayofweek"] = df_clean["tpep_pickup_datetime"].dt.dayofweek
df_clean["pickup_month"] = df_clean["tpep_pickup_datetime"].dt.month
df_clean["pickup_year"] = df_clean["tpep_pickup_datetime"].dt.year

df_clean["total_fare_amount"] = (
    df_clean["total_amount"] -
    df_clean["tip_amount"]
)

df_clean = df_clean[
    (df_clean["trip_distance"] > 0) &
    (df_clean["trip_distance"] < 100) &
    (df_clean["trip_duration_min"] > 1) &
    (df_clean["trip_duration_min"] < 180) &
    (df_clean["total_fare_amount"] > 0) &
    (df_clean["total_fare_amount"] < 300)
].copy()

df_clean_total = df_clean.copy()

print("df_sample:", df_sample.shape)
print("df_clean_total:", df_clean_total.shape)
```

```python
import matplotlib.pyplot as plt

hours = list(hourly_analysis["pickup_hour"])
fares = list(hourly_analysis["total_amount"])

# מוסיפים את הנקודה הראשונה לסוף
hours.append(24)
fares.append(fares[0])

plt.figure(figsize=(12,6))

plt.plot(hours, fares, marker='o')

plt.xticks(range(25))
plt.grid(True)

plt.title("Average Fare by Hour (Cyclic)")
plt.xlabel("Hour")
plt.ylabel("Average Fare ($)")

plt.show()
```

```python
hourly_analysis["hour_sin"] = np.sin(
    2 * np.pi * hourly_analysis["pickup_hour"] / 24
)

hourly_analysis["hour_cos"] = np.cos(
    2 * np.pi * hourly_analysis["pickup_hour"] / 24
)
```

```python
plt.figure(figsize=(8,8))

scatter = plt.scatter(
    hourly_analysis["hour_cos"],
    hourly_analysis["hour_sin"],
    c=hourly_analysis["total_amount"],
    s=250,
    cmap="viridis"
)

for _, row in hourly_analysis.iterrows():
    plt.text(
        row["hour_cos"] * 1.08,
        row["hour_sin"] * 1.08,
        str(int(row["pickup_hour"])),
        ha="center",
        va="center",
        fontsize=10,
        fontweight="bold"
    )

plt.colorbar(scatter, label="Average Fare")

plt.xticks([])
plt.yticks([])

plt.xlabel("")
plt.ylabel("")

plt.axis("equal")

plt.title("Hour of Day as Cyclic Feature")

plt.show()
```

```python
trips_per_hour = (
    df_clean_total
    .groupby("pickup_hour")
    .size()
    .reset_index(name="trip_count")
)

trips_per_hour.head()
```

```python
plt.figure(figsize=(12,5))

plt.bar(
    trips_per_hour["pickup_hour"],
    trips_per_hour["trip_count"]
)

plt.xlabel("Hour of Day")
plt.ylabel("Number of Trips")
plt.title("Trips by Hour")

plt.xticks(range(24))

plt.grid(axis='y')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

hours = trips_per_hour["pickup_hour"]
trips = trips_per_hour["trip_count"]

theta = 2 * np.pi * hours / 24

fig = plt.figure(figsize=(10, 10))
ax = plt.subplot(111, projection="polar")

norm = mpl.colors.Normalize(vmin=trips.min(), vmax=trips.max())
colors = plt.cm.viridis(norm(trips))

ax.bar(
    theta,
    trips,
    width=2 * np.pi / 24,
    color=colors,
    edgecolor="white",
    linewidth=0.8
)

ax.set_theta_zero_location("N")
ax.set_theta_direction(-1)

ax.set_xticks(2 * np.pi * np.arange(24) / 24)
ax.set_xticklabels(range(24), fontsize=11)

ax.set_yticks([])
ax.set_ylim(0, trips.max() * 1.1)

ax.set_title(
    "Trips by Hour (24-Hour Clock)",
    fontsize=16,
    fontweight="bold",
    pad=25
)

ax.grid(alpha=0.25)

sm = plt.cm.ScalarMappable(cmap="viridis", norm=norm)
sm.set_array([])

cbar = plt.colorbar(
    sm,
    ax=ax,
    pad=0.1,
    shrink=0.8
)

cbar.set_label("Number of Trips", fontsize=11)

plt.tight_layout()
plt.show()
```
